##### pie chart

##### stack plot


```python
#showes how all categories build the total
minutes = [1, 2, 3, 4, 5]
p1 = [1, 2, 3, 3, 2]
p2 = [1, 1, 2, 3, 4]
plt.stackplot(minutes, p1, p2, labels=['p1', 'p2'], colors=['#B6B8C1', '#1F3A5F'] )
plt.legend()
plt.show()
```

## learning with the data


```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Online_Retail/Online_Retail.csv"
l_data = pd.read_csv(url, encoding='latin1')
```

```python
l_data.head()
```

```python
from collections import Counter as counter
```

```python
country_counter = Counter(l_data['Country'])
top_5 = [country for country, count in country_counter.most_common(5)]
print(top_5)
```

```python

counts = [count for country, count in country_counter.most_common(5)]
plt.bar(top_5, counts ,label='top 5 countries' )
plt.legend()
```
