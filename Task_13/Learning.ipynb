Series vs data frame?

Series is a one-dimensional labeled array (like a single column)


DataFrame is a 2-dimensional table made up of multiple Series (rows and columns)

```python
# to access multiple columns  
learn__D[['item_name', 'quantity']]
```

iloc vs loc ?

```python
# iloc -> to access by index (integer location)
learn__D.iloc[[0, 1], [0, 1, 2]] # access the first two rows and the first three columns
```

```python
# loc -> to access by label (column name)
learn__D.loc[[0, 1], ['order_id', 'quantity','item_name']] 
```

slicing ? 
buttt here it is inclusive


```python
#inclusive of the last index
learn__D.loc[0:3, 'order_id':'item_name'] 
```
