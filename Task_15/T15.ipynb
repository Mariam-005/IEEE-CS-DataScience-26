### Step 4: Handle Missing Values

##### Spot strings like "UNKNOWN" and "ERROR" that Pandas doesn't see as nulls yet

```python
cafe_DF[(cafe_DF == 'UNKNOWN') | (cafe_DF == 'ERROR')]
```

##### Use `.replace()` to convert "UNKNOWN" and "ERROR" to np.nan

```python
cafe_DF[(cafe_DF == 'UNKNOWN') | (cafe_DF == 'ERROR')]
cafe_DF = cafe_DF.replace(['UNKNOWN', 'ERROR'], np.nan)

```

### Step 5:Text Sanitization

##### remove hidden spaces from all string columns

```python
cafe_DF = cafe_DF.apply(lambda col: col.str.strip() 
    if col.dtype == 'object' else col)
```

##### convert all Item names to lowercase for consistency

```python
cafe_DF = cafe_DF.apply(lambda col: col.str.lower() 
   if col.dtype == 'object' else col)
cafe_DF.head()
```

##### use a dictionary to fix any typos in item names if they exist

```python
print(cafe_DF['Item'].unique())
print(cafe_DF['Payment Method'].unique())
print(cafe_DF['Location'].unique())

#not found
```

### Step 6: Fix Data Types
##### fix incorrect data types (Quantity, Price Per Unit, Total Spent .....)

```python
cafe_DF.info()
```

```python
cafe_DF['Quantity'] = pd.to_numeric(cafe_DF['Quantity'], errors='coerce')
cafe_DF['Price Per Unit'] = pd.to_numeric(cafe_DF['Price Per Unit'], errors='coerce')
cafe_DF['Total Spent'] = pd.to_numeric(cafe_DF['Total Spent'], errors='coerce')
#date 
cafe_DF['Transaction Date'] = pd.to_datetime(cafe_DF['Transaction Date'], errors='coerce')
cafe_DF.info()
```

##### identify and fix rows where Quantity might be zero or negative

```python
cafe_DF.loc[cafe_DF['Quantity'] <= 0, 'Quantity'] = np.nan
```

### Step 7: Handle Missing Values (Part 2: Fill NaNs)

##### fill missing `Price Per Unit`

```python

#fill it with the average
cafe_DF['Price Per Unit'] = cafe_DF['Price Per Unit'].fillna(cafe_DF['Price Per Unit'].mean())
```

```python
cafe_DF['Price Per Unit'].isnull().sum()
```

##### handle other missing values appropriately (like Payment Method or Location)

```python
cafe_DF[['Payment Method', 'Location']].isnull().sum()
```

```python
# can use the most frequent val
cafe_DF["Payment Method"] = cafe_DF["Payment Method"].fillna(
    cafe_DF["Payment Method"].mode()[0]
)

cafe_DF["Location"] = cafe_DF["Location"].fillna(
    cafe_DF["Location"].mode()[0]
)


cafe_DF[['Payment Method', 'Location']].isnull().sum()
```

### Step 8: Logical Validation

##### create a new column `Expected_Total` by multiplying `Quantity * Price Per Unit`

```python
cafe_DF['Expected_Total'] = cafe_DF['Quantity'] * cafe_DF['Price Per Unit']
cafe_DF.head()
```

##### filter rows where Total Spent is NOT equal to Expected_Total

```python
cafe_DF[cafe_DF['Total Spent'] != cafe_DF['Expected_Total']]
```

##### fill missing or incorrect Total Spent values using the Expected_Total calculation.

```python
cafe_DF['Total Spent'] = cafe_DF['Expected_Total']
cafe_DF[['Total Spent', 'Expected_Total']].head()
```

### Step 9: Handle Dates

##### convert Transaction Date to datetime

```python
#already did 
cafe_DF.info()
```

##### identify transactions with dates in the future (if any) and handle them

```python
max_d = cafe_DF['Transaction Date'].max()
cafe_DF.loc[cafe_DF['Transaction Date'] > pd.Timestamp.today(), 'Transaction Date'] = max_d
```

##### extract "Year" and "Month" into separate columns for analysis

```python
cafe_DF['Year'] = cafe_DF['Transaction Date'].dt.year
cafe_DF['Month'] = cafe_DF['Transaction Date'].dt.month
cafe_DF.head()
```
