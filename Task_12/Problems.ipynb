```python
r_arr=np.random.randint(1,101,(5,5))
print(r_arr)
print()

M_row=np.max(r_arr,axis=1)
print(f'max in each row: {M_row}')

S_col=np.sum(r_arr,axis=0)
print(f'sum of each column: {S_col}')

m_arr=np.mean(r_arr)
print(f'mean arr {m_arr}')

bool_arr=np.where(r_arr>50, 1, 0)
print(f'bool arr: {bool_arr}')
```
